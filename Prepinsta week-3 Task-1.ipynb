Question 3: Outliers

```python
def find_outliers(column_name):
    q1 = column_name.quantile(0.25)
    q3 = column_name.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = (column_name < lower_bound) | (column_name > upper_bound)
    return outliers
```

```python
find_outliers(df['Age'])
```

```python
find_outliers(df['Salary'])
```

Question 13: Location Accuracy

```python
import pandas as pd
from difflib import get_close_matches
```

```python
data = {'Location': ['New York', 'India', 'Australia']}
df = pd.DataFrame(data)
```

```python
unique_locations = df['Location'].unique()
print("Unique Locations:")
print(unique_locations)
```

```python
def find_close_matches(location, all_locations):
    close_matches = get_close_matches(location, all_locations, n=1, cutoff=0.8)
    return close_matches[0] if close_matches else None
```

```python
all_locations = unique_locations.tolist()
df['Corrected_Location'] = df['Location'].apply(lambda x: find_close_matches(x, all_locations))
```

```python
print("\nDataFrame with Potential Corrections:")
print(df[['Location', 'Corrected_Location']])
```

Question 14: Handling Categorical Data

```python
data = {'Category': ['Index', 'Age', 'Salary', 'Established', 'Rating', 'Location', 'Easy apply'],
        'Status': ['Active', 'active', 'Active', 'Active', 'active', 'Active', 'Active']}
df = pd.DataFrame(data)
```

```python
df_one_hot = pd.get_dummies(df, columns=['Category', 'Status'], prefix=['Cat', 'Stat'])
```

```python
df_label = df.copy()
for column in df_label.columns:
    df_label[column] = df_label[column].astype('category').cat.codes
```

```python
print("Original DataFrame:")
print(df)
```

```python
print("\nOne-Hot Encoded DataFrame:")
print(df_one_hot)
```

```python
print("\nLabel Encoded DataFrame:")
print(df_label)
```

Question 15: Consistent Rating Scale

```python
data = {'Rating': ['5.4', '3.5', '2.4', '1.0', '7.7']}
df = pd.DataFrame(data)
```

```python
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
```

```python
invalid_ratings = df[df['Rating'].isnull()]
if not invalid_ratings.empty:
    print("Invalid Ratings:")
    print(invalid_ratings)
```

```python
df['Normalized_Rating'] = (df['Rating'] - df['Rating'].min()) / (df['Rating'].max() - df['Rating'].min())
```

```python
print("\nDataFrame with Normalized Ratings:")
print(df[['Rating', 'Normalized_Rating']])
```

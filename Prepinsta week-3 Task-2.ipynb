```python
data = {'Item Name': ['Chips and Fresh Tomato Salsa', 'lzze', 'Nantucket Nectar', 'Chips and Tomatillo-Green Chili salsa', 'Chicken bowl', 'Steak Burrito', 'Chicken Salad Bowl']}
df = pd.DataFrame(data)
```

```python
df_one_hot = pd.get_dummies(df, columns=['Item Name'], prefix='Item')
```

```python
df_label = df.copy()
for column in df_label.columns:
    df_label['Item Name'] = df_label['Item Name'].astype('category').cat.codes
```
